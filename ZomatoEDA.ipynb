```python
df.isnull().sum()
# In Cuisines we have 9 missing values
```

```python
# This is a List Comprehension i.e arranging the columns in the form of a list,
#features(temporary variable) will iterate thorugh df.columns and check if the 'feature' is null and greater than 0
# We directly get the column name which has Null values gt 0, 
# however we do not get the number of MISSING VALUES FOR THE PARTICULAR COLUMN

[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#HEATMAP
sns.heatmap(df.isnull(),yticklabels=False,cbar=False, cmap='viridis')
```

```python
df_country=pd.read_csv('Country-code.csv')
```

```python
df_country.head()
```

```python
# 'How' specifies fow you want to join the datafarames - left join / righ join/inner/outer
# on shows on which variable you want to join the 2 dataframes- SIMILAR TO SQL
# Merge is used to join both the dataframes - df and df_country

final_df=pd.merge(df,df_country, on='Country Code', how='left')
final_df.head()
```

```python
# To check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
# we are here trying to find how many countries' records we have and how many records are there for each country
# OBSERVATIONS:
# MAXIMUM : INDIA
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
# reason behind doing this is to CREATE A PIE CHART
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values,labels=country_names)
```

**THE PIE CHART DOES NOT LOOK NEAT AS THERE ARE TOO MANY COUNTRIES WITH VERY LESS PERCENTAGE, HENCE WE ARE GOING TO SELECT TOP 5 OR TOP 10 COUNTRIES**

```python
#PIE CHART TO SHOW TOP 3 COUNTRIES WHO USE ZOMATO
plt.pie(country_values[:3],labels=country_names[:3], autopct='%1.2f%%')
```

**Observations:
Zomato
Maximum Records or transactions : India, then USA , followed by UK**


```python
final_df.columns
```

```python
# We can see the range of ratings
final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
# Size: gives the count of the ratings
# Reset.index(): sets the index column 
# Rename: Renaming the last column '0' to 'Rating count'. Any column can be renamed
```

**OBSERVATIONS:**
1. whenever the rating is from 4.5-4.9, it indicates the rating is excellent
2. when the ratings are between 4.0-4.4- it is very good
3. when the ratings are between 3.5-3.9- it is good
4. when the ratings are between 2.5-3.4- it is Average
5. when the ratings are between 1.8-2.4- it is Poor
6. Number of orders have not been rated - 2148

```python
ratings
```

```python
final_df['Aggregate rating'].value_counts().sum()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # TO INCREASE OR DECREASE THE SIZE OF THE PLOT
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
# palette is used to give 'Aggregate rating' the color as present in the 'Rating color' column
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```

**OBSERVATIONS:**
1. Maximum number of record- Not rated
2. Maximum number of rating are between 2.5-3.4

```python
# Count plot
# this plot is used to find the count of categorical vqlues
# Count here represents the frequency of the Aggregate rating present in the dataframe, it is NOT THE RATING COUNT.
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
